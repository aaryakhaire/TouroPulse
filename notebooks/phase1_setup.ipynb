## 🧹 2. Deep Data Engineering (The Cleaning Pipeline)

We sanitize the 119k observations to ensure the Dash dashboard remains high-speed and statistically accurate.

### Transformations:
- **Null Mitigation**: Filling children/adr/agent columns with medians/zeros.
- **Strategic Mapping**: Converting `is_canceled` to Dash-compatible string labels (yes/no).
- **Temporal Sorting**: Converting arrival months into categorical time-series objects.

```python
# Copy raw data for sanitization
bookings_cleaned = bookings_raw.copy()

# Handle missing values professionally
bookings_cleaned['children'].fillna(0, inplace=True)
bookings_cleaned['country'].fillna('Unknown', inplace=True)
bookings_cleaned['agent'].fillna(0, inplace=True)
bookings_cleaned['company'].fillna(0, inplace=True)

# Strategic Normalization: ADR (Average Daily Rate)
# Removing outliers (0 or negative ADR is often internal/staff booking not market-relevant)
bookings_cleaned = bookings_cleaned[bookings_cleaned['adr'] > 1]

# Dash Mapping: is_canceled (Boolean to String)
bookings_cleaned['is_canceled_label'] = bookings_cleaned['is_canceled'].map({1: 'yes', 0: 'no'})

# Categorical Month Sorting: Ensures chronological linearity in plots
MONTH_ORDER = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
bookings_cleaned['arrival_date_month'] = pd.Categorical(bookings_cleaned['arrival_date_month'], categories=MONTH_ORDER, ordered=True)

# Export Final Cleaned Dataset
bookings_cleaned.to_csv('../data/cleaned_bookings.csv', index=False)
print("✅ CLEANED: Data synchronized to /data/cleaned_bookings.csv")
```

## 🧠 3. Sentiment Intelligence Engine (Heavy NLP)

We use **Sentiment Polarity Scoring (VADER)** to quantify guest emotional states. This powers the 'Approval Index' in the dashboard.

### Linguistic Keyword Frequency Mapping
We simultaneously extract exactly what guests are talking about regarding operational concerns.

```python
analyzer = SentimentIntensityAnalyzer()

def extract_sentiment_data(text):
    scores = analyzer.polarity_scores(str(text))
    comp = scores['compound']
    if comp >= 0.05: return 'positive', comp
    elif comp <= -0.05: return 'negative', comp
    else: return 'neutral', comp

print("🚀 Processing 20k NLP Guest Feedbacks...")
sentiment_res = reviews_raw['Review'].apply(lambda x: pd.Series(extract_sentiment_data(x)))
reviews_raw[['sentiment_label', 'sentiment_score']] = sentiment_res

# Keyword Pulse Counting (Guest Satisfaction Mapping)
OP_KEYWORDS = ['room', 'staff', 'service', 'clean', 'breakfast', 'location', 'pool', 'bed', 'price', 'wifi']
reviews_raw['lower_review'] = reviews_raw['Review'].str.lower()
all_words = " ".join(reviews_raw['lower_review'])
keyword_counts = {kw: all_words.count(kw) for kw in OP_KEYWORDS}
df_keywords = pd.DataFrame(list(keyword_counts.items()), columns=['Keyword', 'Frequency'])

# Export Sentiment Intelligence
reviews_raw.to_csv('../data/reviews_with_sentiment.csv', index=False)
print("✅ PROCESSED: Sentiment Intelligence mapped to /data/reviews_with_sentiment.csv")
```
